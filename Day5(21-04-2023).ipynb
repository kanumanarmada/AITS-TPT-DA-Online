### Types of plotting:

#### 1. Line plot:
            - It is simple and easy to organize data.
            - plt.plot(array1,array2)

```python
X = [10,12,14,16]
Y = [3,4,5,6]
plt.plot(X,Y)
```

```python
x = [10,12,14,16]
y = [3,4,5,6]
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')
```

```python
# multiple line graphs...
import numpy as np
x = np.linspace(0,5,20)
plt.plot(x,x*2,color='red',label="multiply by 2")
plt.plot(x,x*3,color='blue',label="multiple by 3")
plt.plot(x,x*4,color='green',label="mulitple by 4")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('multiple line graphs')
plt.legend()
```

```python
x1 = np.arange(100)
print(x1)
```

```python
plt.plot(x1,x1+10,color="r",linestyle='--')
plt.plot(x1,x1+15,color='g',linestyle=':')
plt.plot(x1,x1+20,color='y',linestyle='-.')
```

```python
x2 = np.linspace(10,30,50)
print(x2)
```

```python
plt.plot(x2,np.sin(x2),color='r',marker='o',markersize=10,markerfacecolor='w')
plt.plot(x2,np.cos(x2),color='b',marker='D',markersize=10,markerfacecolor='pink')
```

### bar chart:
            - By using bar() function we can draw the bar chart.
            - plt.bar()

```python
year=['2000','2001','2002','2003']
goldprice = [10000,13000,45000,4000]
plt.bar(year,goldprice,color='r')
```

```python
year=[2020,2021,2023]
gold=[23000,45000,58000]
silver=[1000,4500,7800]
plt.bar(year,gold,color='g',label='goldprice')
plt.bar(year,silver,color='pink',label="silverprice")
plt.legend()
```

```python
c=['tcs','wipro','hcl']
s=[12000,24000,10000]
plt.bar(c,s,color=['r','g','y'],edgecolor=['b','y','r'])
```

```python
# scatter plot:
x =[1,2,3]
y=[10,20,30]
plt.scatter(x,y,color='r',marker='*')
```

```python
# pie chart()
subjects=['java','python','dbms','c language']
marks=[45,14,56,78]
plt.pie(marks,labels=subjects,colors=['r','g','b','y'],
        autopct='%1.2f%%',
       shadow=True)
plt.show()
```

```python
# To print the 1 to 100 prime numbers?
# what is prime?
# It is divisible by 1 and itself?
# 2 3 5 7 11
# remainder 2 times zero-> 

for i in range(1,101): # pick the numbers outerforloop
    rem = 0
    for j in range(1,i+1): # iterate that number inner forloop
        if(i%j==0): 
            rem = rem+1
    if rem == 2:
        print(i,end=" ")           
```
